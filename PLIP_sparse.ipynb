```python
#Build indices and values lists for the sparse matrix
```

```python
def indices(mylist, value):
    return [i for i,x in enumerate(mylist) if x==value]
```

```python
lig_index = [0]*2
lig_index
```

```python
prot = []
lig = range(len(lig_list))
lig_index = []
for l in lig:
    list_col = indices(data_list[l][2], 1)
    lig_index += [l]* len(list_col)
    for c in list_col:
        prot += [c+ (l*1155)]
print (len(lig_index), len(prot))
```

```python
fing = np.array([1]*(len(prot)))
row = np.array(lig_index)
col = np.array(prot)
```

```python
#Sparse matrix generation
```

```python
plip_sparse = csr_matrix((fing, (row, col)))
```
